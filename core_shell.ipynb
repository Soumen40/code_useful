```python

## This is for interchange of coordinate (Ni5Ti8 to Ni8Ti5(first 5 coord go to last and last 8 come to first))

### Independent of ordering of elements in POSCAR

from pymatgen.io.vasp import Poscar
from pymatgen.core import Lattice, Molecule, Structure
from copy import deepcopy
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer, PointGroupAnalyzer


ele = ["Ti","V","Cr","Mn","Fe","Co","Cu","Zn"]
ele = ["Sc","Y","Zr","Nb","Mo","Tc","Ru","Rh","Pd","Ag","Cd"]
#ele = ["La","Hf","Ta","W","Re","Os","Ir","Pt","Au","Hg"]

with open("POSCAR",'r') as f1:
    p=[]
    for line in f1:
        p.append(line)
#print(p)

## For box size

a = float(p[2].split()[0])
b = float(p[3].split()[1])
c = float(p[4].split()[2])
#print(a,b,c)

## For number of different atoms

ni = int(p[6].split()[0])
othr = int(p[6].split()[1])

#print(ag,"\n",othr)


### For coordinates of all atoms

p1=deepcopy(p)
#print("\np1",p1)
del p1[0:8]
#print("\np1",p1)
s = []
for j in p1:
    d=[]
    j = j.split()
    #del j[3]  ### This is for if there is element name alongside the coordinates in POSCAR
    
    for i in j:
        d.append(float(i))
    s.append(d)
#print(s)

## New Coordinates after interchanging
s_othr = s[:ni]
s_Ni = s[-othr:]

# # #print(mean)
# for i in range(len(s)):
#     s[i][0]+=-5
#     s[i][1]+=-5
#     #s[i][2]+=-5

# print("all Coordinates:",s)

# #d=[]
# # L1 = [6.641440272,1.3276691,24.834870889]
# # L2 = [6.641440272,4.748766919,24.834870889]

# # #mean = [(L1[i]+L2[j])/2 for i in range(len(L1)) for j in range(len(L2)) if i==j]

# mean = [ 3.70300,   4.98850 ,  8.00000]

# for i in range(len(s)):
#     s[i][0]+=mean[0]
#     s[i][1]+=mean[1]
#     s[i][2]=mean[2] - 4




for j in range(len(ele)):
    spec = ['Ni' for j in range(len(s_Ni))] #+ [ele[j] for j in range(len(s_othr))] 
    for k in range(len(s_othr)):
        spec.append(ele[j])
    coor = s_Ni + s_othr

    lat = [[a,0,0],[0,b,0],[0,0,c]]

    ## For symmetry analysis

    mol_p = Molecule(spec,coor)
    pga_p = PointGroupAnalyzer(mol_p, tolerance=0.000001)
    pg_p = pga_p.get_pointgroup()

    stru = Structure(lat,spec,coor,coords_are_cartesian = False)
    poscar = Poscar(stru)
    #print("poscar2",poscar)
    #Poscar.write_file(poscar,'POSCAR_'+'Ti'+str(ni)+'Ni'+str(othr)+'_'+str(ni)+'_'+str(pg_p))

    Poscar.write_file(poscar,'POSCAR_'+'Ni'+str(len(s_Ni))+ele[j]+str(len(s_othr))+'_'+str(pg_p))


```

```python

## If you have POSCAR this code give you file "test_potcar" to generate POTCAR from vaspkit

## cat test_potcar | vaspkit

def get_POTCAR(POSCAR):
    """
    Input --- POSCAR
    Rteurn ---- VASPKIT sensetive file for POTCAR generation
    """
    POSCAR = "POSCAR"
    with open(POSCAR,'r') as f1:
        p=[]
        for line in f1:
            p.append(line)


    ni = str(p[5].split()[0])
    othr = str(p[5].split()[1])
    f = open("test_potcar",'w')
    f.write("104\n")
    f.write(ni+"\n")

    f.write(othr)
    f.close()
    return f

get_POTCAR("POSCAR")
```

```python
import os
import shutil


def get_POSCAR_in_folder(all_list):
    """
    We have POSCAR files in folder names POSCAR_NixTiy_D1d
    Generate folder name NixTiy and copy this POSCAR_NixTiy_D1d to this folder
    """
    

    
    #print(all_list)

    only_POSCAR = []
    for POSCAR in all_list:
        if POSCAR[:7] == "POSCAR_":
            only_POSCAR.append(POSCAR)
    #print(" POSCAR only present",only_POSCAR,len(only_POSCAR))

    fol_create = []
    for fol in only_POSCAR:
        fol_create.append(fol.split("_")[1])
    #print("folder Created",fol_create)


    for folder in fol_create:
        os.mkdir(folder)

    path = os.getcwd()
    #print("main directory path",path)

    # source_file = path + '/' + only_POSCAR[0]
    # print("source file path",source_file)
    # destination_folder = path + '/' + fol_create[0]
    # print("destiantion folder path",destination_folder)
    # destination_file = os.path.join(destination_folder, source_file)

    # print("destination file path",destination_file)
    # shutil.copy(source_file, destination_folder)

    for j in range(len(only_POSCAR)):
        source_file = path + '/' + only_POSCAR[j]
        destination_folder = path + '/' + fol_create[j]
        #destination_file = os.path.join(destination_folder, source_file)
        # check if destination folder exists
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)

        # copy the file
        shutil.copy(source_file, destination_folder)
    


    
all_list = os.listdir(".")
get_POSCAR_in_folder(all_list)
```
